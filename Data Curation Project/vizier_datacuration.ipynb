```python
df1 = pd.read_csv("Downloads/cleaned__data.csv")
```

```python
df1
```

```python
import numpy as np

#city
#Imputing missing values in the 'city' column based on the 'country' column and using the mode of the city within each specific country.
df1['state'] = df.groupby('country')['state'].transform(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else np.nan))
df1['state'].fillna(df1['state'].mode()[0], inplace=True)

df1.isna().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly
import plotly.graph_objects as go
```

```python

d = df1['country'].value_counts()[0:20]
c_plot = sns.barplot(x=d.index, y=d.values)

# Adding labels and title
plt.xlabel('Country')
plt.ylabel('Number of Billionaires')
plt.title('Top 20 Countries with the Most Billionaires')

c_plot.set_xticklabels(c_plot.get_xticklabels(), rotation=90)
#Container for Showing values
for i in c_plot.containers:
    c_plot.bar_label(i,)
plt.show()
```

```python
#Percentage Count of Self-Made and Inherited Billionaires:

d = df1['selfMade']
d =d.map({True: 'Self Made Billionaires', False: 'Inherited Wealth'})


val = d.value_counts().values 
keys = d.value_counts().index
  
explode = [0, 0.1] 
# define Seaborn color palette to use 
palette_color = sns.color_palette('bright') 
  
# plotting data on chart 
plt.pie(val, labels=keys, colors=palette_color, 
        explode=explode, autopct='%.0f%%') 
  
# displaying chart 
plt.show() 
```

```python
#Percentage Count of Male and Female Billionaires:
d = df1['gender']
d =d.map({'M': 'Male Billionaires', 'F': 'Female Billionaires'})


val = d.value_counts().values 
keys = d.value_counts().index

explode = [0, 0.1] 
# define Seaborn color palette to use 
palette_color = sns.color_palette('dark') 
  
# plotting data on chart 
plt.pie(val, labels=keys, colors=palette_color, 
        explode=explode, autopct='%.0f%%') 
  
# displaying chart 
plt.show() 
```

```python
#Distribution of Each Industry in Total Wealth age vise
sns.histplot(data=df1, x="age", hue = 'industries',kde=True)
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,4))

df_industry = df1['industries'].value_counts().reset_index()
df_industry_worth = df1.groupby('industries')['finalWorth'].sum().reset_index()

## Determine the industry with the most number of billionaires

sns.barplot(data=df_industry, x='count', y='industries', palette=sns.husl_palette(24), ax=axes[0])
axes[0].set_title('Number of Billionaires by Industry')
axes[0].set_xlabel('Number of Billionaires')

## Determine the industry with the most wealth

sns.barplot(data=df_industry_worth.sort_values(by='finalWorth', ascending=False), x='finalWorth', y='industries', palette=sns.husl_palette(24), ax=axes[1])
axes[1].set_title('Distribution of Wealth by Industry')
axes[1].set_xlabel('Worth')

plt.tight_layout()
plt.show()
```

```python
gdp_worth = df1.groupby('countryOfCitizenship')[['finalWorth','gdp_country']].mean().reset_index()
cpi_worth = df1.groupby('countryOfCitizenship')[['finalWorth','cpi_country']].mean().reset_index()
cpi_change_worth = df1.groupby('countryOfCitizenship')[['finalWorth','cpi_change_country']].mean().reset_index()
tax_worth = df1.groupby('countryOfCitizenship')[['finalWorth','tax_revenue_country_country']].mean().reset_index()
total_tax_worth = df1.groupby('countryOfCitizenship')[['finalWorth','total_tax_rate_country']].mean().reset_index()

fig, axes = plt.subplots(2,3, figsize=(12,5))

## Plot the relationship between the country's GDP and the mean net worth of billionaires from that country
sns.scatterplot(data=gdp_worth, x='gdp_country', y='finalWorth', ax=axes[0,0])
axes[0,0].set_xlabel('GDP')
axes[0,0].set_ylabel('Mean Net Worth by Country')

## Plot the relationship between the country's CPI and the mean net worth of billionaires from that country
sns.scatterplot(data=cpi_worth, x='cpi_country', y='finalWorth', ax=axes[0,1])
axes[0,1].set_xlabel('CPI')
axes[0,1].set_ylabel('Mean Net Worth by Country')

## Plot the relationship between the country's CPI change and the mean net worth of billionaires from that country
sns.scatterplot(data=cpi_change_worth, x='cpi_change_country', y='finalWorth', ax=axes[0,2])
axes[0,1].set_xlabel('CPI Change')
axes[0,1].set_ylabel('Mean Net Worth by Country')

## Plot the relationship between the country's tax revenue and the mean net worth of billionaires from that country
sns.scatterplot(data=tax_worth, x='tax_revenue_country_country', y='finalWorth', ax=axes[1,0])
axes[1,0].set_xlabel('Tax Revenue')
axes[1,0].set_ylabel('Mean Net Worth by Country')

sns.scatterplot(data=total_tax_worth, x='total_tax_rate_country', y='finalWorth', ax=axes[1,1])

plt.tight_layout()
plt.show()
```

```python
tert_worth = df1.groupby('countryOfCitizenship')[['finalWorth','gross_tertiary_education_enrollment']].mean().reset_index()
prim_worth = df1.groupby('countryOfCitizenship')[['finalWorth','gross_primary_education_enrollment_country']].mean().reset_index()

fig, axes = plt.subplots(1,2, figsize=(8,3))

## Plot the relationship between the country's Gross Tertiary Education Enrollment and the mean net worth of billionaires from that country
sns.scatterplot(data=tert_worth, x='gross_tertiary_education_enrollment', y='finalWorth', ax=axes[0])
axes[0].set_xlabel('Gross Tertiary Education Enrollment')
axes[0].set_ylabel('Mean Net Worth by Country')

## Plot the relationship between the country's Gross Primary Education Enrollment and the mean net worth of billionaires from that country
sns.scatterplot(data=prim_worth, x='gross_primary_education_enrollment_country', y='finalWorth', ax=axes[1])
axes[1].set_xlabel('Gross Primary Education Enrollment')
axes[1].set_ylabel('Mean Net Worth by Country')

plt.tight_layout()
plt.show()
```
